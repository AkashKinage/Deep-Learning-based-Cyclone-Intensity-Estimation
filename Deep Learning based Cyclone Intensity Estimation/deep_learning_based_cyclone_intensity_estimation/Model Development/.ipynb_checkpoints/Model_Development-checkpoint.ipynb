## Data Shape

```python
from sklearn.model_selection import train_test_split

# Assuming you have numpy arrays of images (X) and labels (y)
# X should have shape (num_samples, image_height, image_width, num_channels)
# y should have shape (num_samples,)

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
y_train = y_train.reshape(len(y_train), 1)
y_test = y_test.reshape(len(y_test), 1)
```

```python
# Print the shapes of the training and testing sets to verify
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
X_train = X_train/255.0
```

## Model Building

```python
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, LSTM, Dense, BatchNormalization, Dropout, SimpleRNN

def cnn_rnn_model(image_shape, num_classes):
    # Define CNN block for feature extraction
    inputs = keras.Input(shape=image_shape)
    x = Conv2D(64, (3, 3), activation='relu', padding='same')(inputs)
    x = BatchNormalization(axis=1)(x)
    x = MaxPooling2D(pool_size=2)(x)
    
    x = Conv2D(128, (3, 3), activation='relu')(x)
    x = BatchNormalization(axis=1)(x)
#     x = MaxPooling2D(pool_size=2)(x)
    
    x = Conv2D(256, (3, 3), activation='relu')(x)
    x = BatchNormalization(axis=1)(x)
#     x = MaxPooling2D(pool_size=2)(x)
    
    x = Conv2D(512, (3, 3), activation='relu')(x)
    x = BatchNormalization(axis=1)(x)
    x = MaxPooling2D(pool_size=2)(x)
    
    x = Conv2D(512, (3, 3), activation='relu')(x)
#     x = BatchNormalization(axis=1)(x)
#     x = MaxPooling2D(pool_size=2)(x)
    
    x = Flatten()(x)
    x = Dropout(0.3)(x)
    
    x = Dense(64, activation='relu')(x)
#     x = Dense(16, activation='relu')(x)
    x = Dense(1, activation=None)(x)

    # Reshape for RNN input (samples, timesteps, features)
    x = tf.expand_dims(x, axis=1)  # Add a timestep dimension

    # Define RNN block for sequence modeling
    x = SimpleRNN(128, return_sequences=True)(x)  # Analyze features across timesteps
    x = SimpleRNN(64, return_sequences=True)(x)  # Extract higher-level features
#     x = SimpleRNN(64, return_sequences=True)(x)
#     x = SimpleRNN(32)(x)
    # x = Dropout(0.5)(x)
    x = Dense(64, activation='relu')(x)
#     x = Dense(16, activation='relu')(x)
    x = Dense(16, activation='relu')(x)

    # Final classification layer
    outputs = Dense(num_classes, activation=None)(x)

    model = tf.keras.Model(inputs=inputs, outputs=outputs)
    return model

# Replace with your data shapes and number of classes
model = cnn_rnn_model((50, 50, 1), 1)  # Assuming images are 28x28x3 (RGB) and 3 cyclone classes

model.compile(
        optimizer='adam',
        loss='mse',
        metrics=[metrics.MeanAbsoluteError(), metrics.RootMeanSquaredError()]
    )
model.fit(X_train, y_train, epochs=20, validation_data=(X_test, y_test))
```

```python
model.evaluate(X_test, y_test)
```

```python
# Saba
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, LSTM, Dense, BatchNormalization, Dropout, SimpleRNN
from tensorflow.keras import metrics

def cnn_rnn_model(image_shape, num_classes):
    # Define CNN block for feature extraction
    inputs = tf.keras.Input(shape=image_shape)
    x = Conv2D(32, (3, 3), activation='relu')(inputs)
    x = BatchNormalization()(x)
    x = MaxPooling2D(pool_size=(2, 2))(x)
    
    x = Conv2D(64, (3, 3), activation='relu')(x)
    x = BatchNormalization()(x)
    x = MaxPooling2D(pool_size=(2, 2))(x)
    
    x = Conv2D(256, (3, 3), activation='relu')(x)
    x = BatchNormalization()(x)
    x = MaxPooling2D(pool_size=(2, 2))(x)
    
    x = Flatten()(x)
    x = Dropout(0.5)(x)
    x = Dense(256, activation='relu')(x)

    # Reshape for RNN input (samples, timesteps, features)
    x = tf.expand_dims(x, axis=1)  # Add a timestep dimension

    # Define RNN block for sequence modeling
    x = SimpleRNN(128, return_sequences=True)(x)  # Analyze features across timesteps
    x = SimpleRNN(64)(x)  # Extract higher-level features
    x = Dense(64, activation='relu')(x)

    # Final classification layer
    outputs = Dense(num_classes, activation=None)(x)

    model = tf.keras.Model(inputs=inputs, outputs=outputs)
    return model

# Replace with your data shapes and number of classes
model = cnn_rnn_model((50, 50, 1), 1)  # Assuming images are 50x50x1 and regression problem with 1 output

model.compile(
        optimizer='adam',
        loss='mse',
        metrics=[metrics.MeanAbsoluteError(), metrics.RootMeanSquaredError()]
    )

# Assuming X_train, y_train, X_test, y_test are defined
history = model.fit(X_train, y_train, epochs=100, validation_data=(X_test, y_test))
```
